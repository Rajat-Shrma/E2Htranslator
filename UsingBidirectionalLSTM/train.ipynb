```python
english_lines = []
hindi_lines = []
for line in lines:
    eng, hin = line.strip().split('\t')
    english_lines.append(eng)
    hindi_lines.append(hin)
    
```

```python
data=pd.DataFrame({'English':english_lines,'Hindi':hindi_lines})
```

```python
import keras
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python

tokenizer_eng = Tokenizer()
tokenizer_hin = Tokenizer() 
tokenizer_eng.fit_on_texts(english_lines)
tokenizer_hin.fit_on_texts(hindi_lines) 
```

```python
tokenizer_eng.word_index
```

```python
tokenizer_hin.word_index
```

```python
data['English']=data['English'].apply(lambda x: x.lower())
data['Hindi']=data['Hindi'].apply(lambda x: x.lower())  
```

```python
data.head()
```

```python
X=tokenizer_eng.texts_to_sequences(data['English'])
y=tokenizer_hin.texts_to_sequences(data['Hindi'])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_len_eng = max(len(seq) for seq in X)
```

```python
max_len_eng
```

```python
max_len_hin = max(len(seq) for seq in y)
```

```python
max_len_hin
```

```python
max_len_hin = 25
max_len_eng = 25
```

```python
X_padded = pad_sequences(X, maxlen=max_len_eng, padding='post')
y_padded = pad_sequences(y, maxlen=max_len_hin, padding='post')
```

```python
X_padded
```

```python
y_padded
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional, TimeDistributed, Dropout
```

```python
english_vocab_size = len(tokenizer_eng.word_index) + 1
hindi_vocab_size = len(tokenizer_hin.word_index) + 1
```

```python
hindi_vocab_size
```

```python
model=Sequential(
    [
    Embedding(input_dim=english_vocab_size, output_dim=128, input_length=max_len_eng),
    Bidirectional(LSTM(256, return_sequences=True)),
    TimeDistributed(Dense(1024, activation='leaky_relu')),
    Dropout(0.5),
    TimeDistributed(Dense(hindi_vocab_size, activation='softmax'))

])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy', patience=5, restore_best_weights=True)
```

```python
history=model.fit(X_padded, y_padded, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```
